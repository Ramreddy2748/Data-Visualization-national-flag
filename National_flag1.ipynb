## Import the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

## Plot  the axis and Plot the rectangle in the figure

```python
fi,axis=plt.subplots(figsize=(6,4))
axis.add_patch(patches.Rectangle((0,4/5),1,1/5,color='orange'))
axis.add_patch(patches.Rectangle((0,3/5),1,1/5,color='white'))
axis.add_patch(patches.Rectangle((0,2/5),1,1/5,color='green'))
```

## Plot the ashoka chakra(circle) of axis(x=0.5,y=0.7) and radius=0.1

```python
circle=plt.Circle((0.5,0.7),0.1,color='blue',fill=False)
axis.add_patch(circle)
c2=plt.Circle((0.5,0.7),0.02,color='black',fill=False)
axis.add_patch(c2)
fi
```

## Plot the "Spokes"

```python
for i in range(0,360,15):
  ang = i * (3.14159265358979323846 / 180)
  x1 = 0.5 + 0.1 * np.cos(ang)
  y1 = 0.7 + 0.1 * np.sin(ang)
  x2 = 0.5 + 0.02 * np.cos(ang)
  y2 = 0.7 + 0.02 * np.sin(ang)
  axis.plot([x1, x2], [y1, y2], 'blue')
fi
```

##   Set the limits for the axis and Label for the axis

```python
axis.set_xlim(0,1)
axis.set_ylim(0,1)
axis.set_xlabel('X')
axis.set_ylabel('Y')
fi
```
